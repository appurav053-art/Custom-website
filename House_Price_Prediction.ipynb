**Imports**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import sklearn.datasets
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor
from sklearn import metrics
```

Importing the House Price Dataset

```python
house_price_dataset = sklearn.datasets.fetch_california_housing()
```

```python
print(house_price_dataset)
```

```python
# Loading the dataset to a pandas dataframe
house_price_dataframe = pd.DataFrame(house_price_dataset.data, columns = house_price_dataset.feature_names)
```

```python
house_price_dataframe.head()
```

```python
# add the target column to the dataframe
house_price_dataframe['price'] = house_price_dataset.target
```

```python
house_price_dataframe.head()
```

```python
# checking the number of rows and columns in the dataframe
house_price_dataframe.shape
```

```python
# check for missing values
house_price_dataframe.isnull().sum
```

```python
# statistical measures of the dataset
house_price_dataframe.describe()
```

Understanding the **correlation** between various features in the dataset

1. Positive Correlation
2. Negative Correlation

```python
correlation = house_price_dataframe.corr()
```

```python
# constructing a heatmap to understand the correlation

plt.figure(figsize=(10,10))
sns.heatmap(correlation, cbar=True, square=True, fmt='.1f', annot=True, annot_kws={'size':8}, cmap='Blues')
```

Splitting the data and target

```python
X = house_price_dataframe.drop(['price'], axis=1)
Y = house_price_dataframe['price']
```

```python
print(X,Y)
```

Splitting the data into training data and test data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.2, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

Model Training

XGBoost Regressor

```python
# load the model
model = XGBRegressor()
```

```python
#training the model with X_train
model.fit(X_train, Y_train)
```

Evaluation

Prediction on training data

```python
# accuracy for prediction on training data
training_data_prediction = model.predict(X_train)
```

```python
print(training_data_prediction)
```

```python
# R Squared Error
score_1 = metrics.r2_score(Y_train, training_data_prediction)

# Mean Absolute Error
score_2 = metrics.mean_absolute_error(Y_train, training_data_prediction)

print('R Sqaured Error:', score_1)
print('Mean Absolute Error:', score_2)
```

Visualize the actuale prices and predicted prices

```python
plt.scatter(Y_train, training_data_prediction)
plt.xlabel("Actual Price")
plt.ylabel("Predicted Price")
plt.title("Actual Price vs Predicted Price")
plt.show()
```

Prediction on test data

```python
# accuracy for prediction on test data
test_data_prediction = model.predict(X_test)
```

```python
# R Squared Error
score_1 = metrics.r2_score(Y_test, test_data_prediction)

# Mean Absolute Error
score_2 = metrics.mean_absolute_error(Y_test, test_data_prediction)

print('R Sqaured Error:', score_1)
print('Mean Absolute Error:', score_2)
```

```python
# Plotting feature importance
plt.figure(figsize=(10, 6))
feat_importances = pd.Series(model.feature_importances_, index=X.columns)
feat_importances.nlargest(10).plot(kind='barh', color='skyblue')
plt.title('Top Features Driving House Prices')
plt.xlabel('Relative Importance Score')
plt.show()
```

```python
# Calculating residuals
residuals = Y_test - test_data_prediction

plt.figure(figsize=(10, 6))
sns.histplot(residuals, kde=True, color='purple')
plt.axvline(x=0, color='red', linestyle='--')
plt.title('Distribution of Prediction Errors (Residuals)')
plt.xlabel('Error (Actual - Predicted)')
plt.ylabel('Frequency')
plt.show()
```

```python
# A more optimized version of the model
model = XGBRegressor(
    n_estimators=1000,
    learning_rate=0.05,
    max_depth=6,
    n_jobs=-1
)
model.fit(X_train, Y_train)
```

```python
# Create a dataframe to compare predictions
comparison_df = pd.DataFrame({'Actual': Y_test, 'Predicted': test_data_prediction})
comparison_df['Error'] = comparison_df['Actual'] - comparison_df['Predicted']

# Show the 5 houses with the biggest prediction errors
print("Top 5 Biggest Errors:")
print(comparison_df.sort_values(by='Error', ascending=False).head())
```

```python
# Calculating distance from Los Angeles (34.05, -118.24)
house_price_dataframe['dist_to_LA'] = np.sqrt((house_price_dataframe['Latitude'] - 34.05)**2 +
                                              (house_price_dataframe['Longitude'] + 118.24)**2)

# Calculating distance from San Francisco (37.77, -122.41)
house_price_dataframe['dist_to_SF'] = np.sqrt((house_price_dataframe['Latitude'] - 37.77)**2 +
                                              (house_price_dataframe['Longitude'] + 122.41)**2)
```

```python
import plotly.express as px

# Map of the errors (Blue = Good prediction, Red = Big error)
fig = px.scatter_mapbox(comparison_df.join(X_test),
                        lat="Latitude", lon="Longitude",
                        color="Error", size="Actual",
                        color_continuous_scale=px.colors.cyclical.IceFire,
                        size_max=15, zoom=5,
                        mapbox_style="carto-positron")
fig.show()
```

```python
def predict_my_house(income, age, rooms, bedrooms, pop, occup, lat, lon):
    input_data = pd.DataFrame([[income, age, rooms, bedrooms, pop, occup, lat, lon]],
                              columns=X.columns)
    prediction = model.predict(input_data)
    print(f"Estimated House Value: ${prediction[0] * 100000:,.2f}")

# Example: 8.0 income, 15 years old, 6 rooms...
predict_my_house(8.0, 15, 6, 1, 300, 2.5, 37.8, -122.2)
```
